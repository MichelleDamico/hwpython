```python
trivia_list = {'How many U.S. States begin with the letter A? ': ['four', '4'],

'Who sings the song that begins: Ground Control to Major Tom? ': ['Bowie', 'David Bowie'],

'Fill in the blank: President William Howard ______. ' : 'Taft',

'How many wives did Ernest Hemingway have?' : ['four', '4'],

'What is the New York State motto? ': 'Excelsior',

'Color blindness means that you have the inability to distinguish between which two colors? ':
['red and green', 'green and red', 'red & green', 'green & red', ' red green', 'green red'],

'In our solar system, which planet has the shortest day? ': 'Jupiter',

'Name one of the 12 people who are on the EGOT recipient list (have one a Emmy, Grammy, Oscar, AND Tony)? ':
['Richard Rogers', 'Helen Hayes', 'Rita Moreno', 'John Gielgud', 'Audrey Hepburn', 'Marvin Hamlish', 'Jonathan Tunkick',
'Mel Brooks', 'Mike Nichols', 'Whoopi Goldberg', 'Scott Rudin', 'Robert Lopez'],

'Which element appears as Pb on the Periodic Table of Elements? ': 'Lead',

'Dr. Frank-N-Furter steals the show in which 1975 cult movie? ': ['Rocky Horror', 'Rocky Horror Picture Show']}

def ask_trivia_question(xquestion, ylist):

    answer = input('\nHERE IS YOUR QUESTION:\n {} \n'.format(xquestion)).title()
    s = 0

    if answer in ylist:
        print('\nYou are correct! The answer is: {}'.format(answer))
        s = 1
    else:
        print('\nYou are incorrect. The answer is: {}'.format(ylist))
    return s

def play_again():

    decision = input('Would you like to play again? Y/N ').title()

    if decision == 'Y':
        print('\nGREAT! We will play again.')
        return True
    else:
        print('\nAll Done! Thanks for playing!\n')
        return False

def main():
    score = 0
    number_of_quizes = 0
    keep_playing = True
    while keep_playing:
        for question, answer_list in trivia_list.items():
            score = score + ask_trivia_question(question, answer_list)
            print('\nYour current score is: {}'.format(score))
        print('\nCongrats! You answered {} out of {} correctly.\n'.format(score, len(trivia_list)))
        number_of_quizes = number_of_quizes + 1
        keep_playing = play_again()
    print('Thanks for playing {} quizes! '.format(number_of_quizes))

#=================================
#let's begin the QUIZ!

main()
```
